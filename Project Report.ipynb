The Elsanow Product Analyst Challenge provided city population data on different countries around the world for an analysis. Here's the link to the Google data studio dashboard I created: https://datastudio.google.com/s/kWeIXw7IUw8


the first step I took was to clean up the dataset and prepare it for analysis and as a data source for the dashboard. I deleted all rows with null values. These were mostly rows that contained value footnotes. Once I had verified that no missing values existed, I created a new column called Continents that mapped all countries to their respective continents. Once this was done, I commenced my analysis.

I used Google Data Studio to create my interactive dashboard. The metrics I used were:

1. Class Distribution of Continents contained in the dataset.
2. Record Counts by Country.
3. Class Distribution by Sex.
4. Interactive map of Population by Country

These metrics can further be filtered by SEX, YEAR and COUNTRY OR AREA.

## Findings

1. The data is not evenly distributed amongst different years and amongst the different continent. As such no meaningful continent level comparison can be made. For example 1998 numbers are gotten solely from Asian countries and 71% of all numbers in 2006 come from Europe
